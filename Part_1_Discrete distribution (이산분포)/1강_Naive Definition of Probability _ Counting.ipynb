### Naive definition of prob.  <br>

"This is the very naive definition of probability,  
so you can only use this definition when you have strong justification for doing so." <br><br><br>

#### Definition : <br>

The probability of an event A is just the rate of the size of the sample space in the denominater and how many of those outcomes did A occur in the numerator. <br><br>

$$ P(A) = \frac {\text{favorable outcomes (to A)}} {\text{Possible outcomes}} $$ <br>

```
We're imagining we have some experiment that we're considering, we have the sample space, we have a subset we're interested in, A. Because we want to know what's 'the chance', what are 'the odds' that some particular event will occur. That's the question we're going to be considering throughout this course.
```
<br><br>

#### Example : flip coin twice <br>

```
What's the probability that both tosses are tails(T)? ... 1/4  
H / H  
H / T  
T / H  
T / T  
according to the naive definition of probability ... 1/4.  

Notice.  
Is it a fair coin?  ... What does it actually mean for a coin to be fair?  
We have to be careful of some circularity here.  
'fair coin'means heads and tails are equally likely!
```
<br><br>

#### Assumption : "fair coin" <br>

- __"all outcomes are equally likely ("동등한 경향성")"__ <br><br>
    that's a reasonable assumption in some problems where we have some kind of symmetry. <br><br>
- __"all outcomes are finitely many outcomes (finite sample space)"__ <br><br>
    if the outcome of the experiment could be like any real number or any integer, <br>
    then the denominater would be infinity and then the above definition is meaningless. <br><br><br>

#### Counter-Example : <br>
```
What's the probability that there's life on Neptune? 
... 1/2
What's the probability that there's intelligent life on Neptune? 
... 1/2 (?)
```
<br>

People are often taking seriously arguments, using the naive definition with no justification.
There's no justification for using the naive definition in that case! <br><br>

There's something severely wrong with - shouldn't it be strictly less likely that there's intelligent life than that there's any kind of life. Or there should be a strict inequality there. That's not reflected. <br><br>

So we'll quickly need to go beyoun this naive definition (in this example case) <br><br><br>

_"However this 'Naive definition of probability' is still importnat."_ <br>
_"It's important both to understand how the subject evolved, but it's also important for a lot of problems where we are able to assume equally likely. I'm just emphasizeing the fact that you need some justification or be very clear about what you're able to assume. And if you are able to assume that all the outcomes are eqaully likely and there's finitely many, then this is the definition of probability, it's perfectily good, otherwise you can't."_

### Multiplication rule <br>

#### Note : Binomial Coefficient (이항계수) <br>


> __Definition__ <br><br>
It's pronounced _'n choose k'_. <br>
Number of a subeset of size k, of group of n people, where order doesn't matter. <br>

$$ \binom {n} {k} = \frac {n!} {(n-k)! \; k!} \qquad , \; 0 \quad if \quad  k > n $$ 

<br>

> __Proof__ <br><br>
$\rightarrow$ "following from multiplication rule."

$$ \frac {n \; (n-1) \; (n-2) \; \cdots \; (n-k+1)} {k!} \; = \; \frac {n!} {(n-k)! \; k!} $$
