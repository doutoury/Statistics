>But certain couting probalems and for certain problems in physics, they are so completely indistinguishable that you can't even think of them as labeled. ### Basic structure <br><br>


_'The general definition of probability'_ <br>
> Up until now, we were assuming __equally likely outcomes__. And we want to go beyond that. We don't want to assume everything is equally likely, and we don't want to have to assume that there are __only finitely many possible outcomes__.

<br> 

For the non-naive definition, we need the notion of __a probability (sample) Space__. <br><br>


#### Definition : Non-naive def. of prob. <br>


A probability space consists of two ingredients, which are $S : sample \; space$ and $P : fuction$. The $function \; P$ takes an event $A \subseteq S$ as input, returns $P(A) \in [0, 1]$. <br><br>

- $S : sample \; space$ <br>
That's the set of all possible outcomes of some experiment. <br>
(from now here, we __not__ have to assume _'equally likely'_ and _'finite'_ for outcomes.) <br>

- $P : function$ * <br>
The domain of $P$ is ' $all \; subsets \; of \; S$ '. <br>
It's a __mapping function__ that takes an $event \; A$ as an $input$. An $event \; A$ is a subset of $S$, $A \subseteq S$, and returns $P(A) \in [0, 1]$. <br>
By standard convention of probability, when the $input$ is an $event \; A$, the $output$ is a $number  \;between \; 0 \; and \; 1$. <br>
<br>
But it's not the kind of function that you usually see, like $f(x) = x^{2}$. <br>


<br><br>
#### * Note : $function$ in Set theory <br>
확률에서의 함수 P 는 '집합론 에서의 함수'. '대수론 에서의 함수' 와 따로 정의 된다. 집합들에 대해서 적용되는 함수관계. <br><br><br>




### Axiom for Non-niave def. of prob. <br><br><br>


What axioms does this $function \; P$ have to satisfy? <br>
What properties do we need $P$ to satisfy, aside from the fact that $P(A)$ is between [0, 1]? <br><br>

you might think that a probability is a very complicated thing like understanding uncertainty is a very complicated thing. But actually we only need 2 axioms (2 rules). <br><br><br>


1. $P(\varnothing) = 0$ &emsp; &emsp; : the probability of the empty set = 0 <br>
$P(S) = 1$ &emsp; &emsp; : the probability of the full space = 1 <br><br>
>that makes sense intuitively, by __Ven diagram__. <br>
>$S$ : the universe of all possible outcomes <br>
>$A$ : the subset of all possible outcomes <br> 
>$s_0$ : an example of an outcome <br><br>
>
>$\lceil \quad S \qquad \qquad \quad \; \rceil$ <br>
>$| \qquad \; \; \lceil \; \; A \qquad \; \rceil \; | $ <br>
>$| \qquad \; \; | \qquad s_0 \; \; \; | \; \, |$ <br>
>$| \qquad \; \; \lfloor \qquad \quad \; \; \rfloor \; |$ <br>
>$\lfloor \qquad \qquad \qquad \quad \rfloor $ <br><br>
>
>So our picture is, before we do the experiment, we just have this general sample sapce. After do the experiment, then we get to observe the outcome. <br>
>And suppose that the actual outcome of the experiment was $s_0$. Now, if that actual outcome is an element of $A$, then we say that _"$A$ occurred"_. And if this $s_0$ is outside of $A$, we say _"it doesn't occur"_. <br><br>
>Now, what would it mean for the empty set to occur? It would mean that this $s_0$ is in the empty set. But nothing is in the empty set. It's empty. So that's why I want this $P(\varnothing)$ to be 0. What that's saying is we want impossible events to have probability 0. <br><br>
>On the other hand, This $S$ itself happens with certainty. If somehow $s_0$ fell outside of that rectangle, that would mean you had the wrong rectangle, $S : sample \; space$. This set $S$ is the whole universe here. So it cant's be outside. So that $P(S)$ is 1. That's basically convention.

<br><br>

The most important axiom is the 2nd one. <br><br>


2. $P(\bigcup_{n=1}^{\infty} A_n) = \sum\limits_{n=1}^{\infty} P(A_n)$ &emsp; &emsp; &emsp; , if $A_1, A_2, ... , A_n$ are __disjoint__ (non-overlapping) <br>
>the probability of the union. (This $ $ is called a __countable union__, where I'm taking a union of infinitely many, called a countably infinitely many. <br><br>
>
>So the probability of the union equals the sum of the probabilities. And there's an important condition there. "That's true if the $A_1, A_2, ..., A_n$ are disjoint, which means that no two of them overlap (non-overlapping).

<br><br>

So those are _"the two axioms of probability"_. <br>

__" Amazingly, from these 2 simple rules, we can derive every single theorem and result in probability. "__
